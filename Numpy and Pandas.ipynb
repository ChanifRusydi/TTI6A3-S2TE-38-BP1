```python
import pandas as pd
```

```python
data = pd.read_csv('epl-goalScorer(20-21).csv')
data.head(5)
```

```python
data.tail(5)
```

```python
# 10 player scores most goal
data.sort_values('goals', ascending=False).head(10)
```

```python
# 10 with longest time 
data.sort_values('time', ascending=False).head(10) 
```

```python
data.sort_values('assists', ascending=False).head(10) 
```

```python
manchester_city = data[data['team_title'] == 'Manchester City']
```

```python
manchester_city.sort_values('goals', ascending=False).head(4)
```

```python
most_goal = data.sort_values('goals', ascending=False).head(1)
most_goal['team_title']
```

```python
most_assist = data.sort_values('assists', ascending=False).head(1)

most_assist
```

```python
kartu_kuning, kartu_merah = most_assist['yellow_cards'], most_assist['red_cards']
kartu_kuning, kartu_merah
```

```python
# player who have both most assist and goal
data[(data['goals'] == most_goal['goals'].values[0]) & (data['assists'] == most_assist['assists'].values[0])]
```

```python
data.groupby('goals').sum().sort_values('goals', ascending=False).head(5)
```

```python
no_card_player = data[(data['yellow_cards'] == 0) & (data['red_cards'] == 0)]
no_card_player.sort_values('goals', ascending=False).head(5)
```
